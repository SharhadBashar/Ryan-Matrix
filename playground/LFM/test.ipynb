```python
import csv
import pandas as pd
from tqdm import tqdm
```

```python
def users(filename):
    data = {
        'user_id': [],
        'country': [],
        'age': [],
        'gender': []
    }     
    with open(filename, 'r', encoding = 'utf8') as f:                 
        reader = csv.reader(f, delimiter = '\t')
        next(reader)
        for row in tqdm(reader):
            # if(row[1] == '' or int(row[2]) < 0 or row[3] not in ['m', 'f']): continue
            data['user_id'].append(row[0])
            data['country'].append(row[1])
            data['age'].append(row[2])
            data['gender'].append(row[3])
    f.close()
    df = pd.DataFrame.from_dict(data)
    # df.to_csv(filename.replace('txt', 'csv')) 
    return df
```

```python
df = users('../../data/LFM/txt/LFM-1b_users.txt')
df.head()
```

```python
df = df[~df['gender'].isin(['m', 'f'])]
df = df[df['country'] != '']
```
