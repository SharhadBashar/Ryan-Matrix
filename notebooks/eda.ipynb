```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme()
sns.set(rc = {'figure.figsize':(20, 9)})
```

```python
df_age = pd.read_csv('../data/table/df_age_NL_table.csv')
df_gender = pd.read_csv('../data/table/df_gender_NL_table.csv')
```

```python
ax = sns.barplot(df_age, x = 'MinAge', y = 'Percentage', hue = 'ContentFormatId')
sns.move_legend(ax, "upper left", bbox_to_anchor =(1, 1))
plt.title('Category distribution by Age in NL')
plt.xlabel('Age Range')
ax.set_xticklabels(['0 - 24', '25 - 34', '35 - 44', '45 - 54', '55 - 64', '65+'])
plt.show()
```

```python
ax = sns.barplot(df_gender, x = 'IsMale', y = 'Percentage', hue = 'ContentFormatId')
sns.move_legend(ax, "upper left", bbox_to_anchor = (1, 1))
plt.title('Category distribution by Gender in NL')
plt.xlabel('Gender')
ax.set_xticklabels(['Female', 'Male'])
plt.show()
```
